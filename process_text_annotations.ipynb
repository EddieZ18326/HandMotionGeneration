```python
from openpyxl import load_workbook
from openpyxl.utils import column_index_from_string
import os

# === Configuration ===
source_file = "/Users/eddie/Downloads/VLMTrain/VLMAnnotations_Train_Recognition.xlsx"
target_file = "/Users/eddie/Downloads/Traj_Train.xlsx"
match_column = "B"  # Column letter to check (e.g., "B" for 2nd column)
# match_value = "Apple"  # The value we want to match

# === Load workbooks ===
source_wb = load_workbook(source_file)
target_wb = load_workbook(target_file)

source_ws = source_wb.active
target_ws = target_wb.active
dirs = os.listdir("/Users/eddie/Trauma_THOMPSON/Trauma_THOMPSON/Hand Tracking Dataset/Hand Tracking Dataset/train/")
col_index = column_index_from_string(match_column) - 1

# === Iterate through rows in the source file ===
for row in source_ws.iter_rows(min_row=2, values_only=False):  # skip header if needed
    # print(row[col_index].value)
    if row[col_index].value in dirs:
        # Append the entire row to the target worksheet
        target_ws.append([cell.value for cell in row])

# === Save target workbook ===
target_wb.save(target_file)

# print(f"Rows with {match_column} = '{}' copied to {target_file}")
```

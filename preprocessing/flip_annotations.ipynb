```python
import os
import json
import copy

def reflect_coco_annotations_in_place(file_path):
    with open(file_path, 'r') as f:
        coco = json.load(f)

    flipped = copy.deepcopy(coco)

    image_dims = {img['id']: (img['width'], img['height']) for img in flipped['images']}

    for ann in flipped['annotations']:
        image_id = ann['image_id']
        width, height = image_dims[image_id]

        x, y, w, h = ann['bbox']
        ann['bbox'][0] = width - x - w
        ann['bbox'][1] = height - y - h

    with open(file_path, 'w') as f:
        json.dump(flipped, f)

def process_all_jsons_in_dir(directory):
    for filename in sorted(os.listdir(directory)):
            file_path = os.path.join(directory, filename)
            print(f"Processing {filename}")
            reflect_coco_annotations_in_place(file_path)

video = "P03_11"
process_all_jsons_in_dir("/Users/eddie/Trauma_THOMPSON/Trauma_THOMPSON/Hand Tracking Dataset/Hand Tracking Dataset/train/" + video + "/" + video+"/")
```
